### mpl

* Matplotlib is the lowlevel library that dominates the ecosystem
* A quick [history and architecture](https://www.aosabook.org/en/matplotlib.html)
* Matplotlib renders a chart in the 'backend' and sends an image to the frontend
* This is why interactivity is hard in Matplotlib (?)

* Seaborn builds on top of Matplotlib enabling more complex plots and better styles
* Bokeh enables interactivity on top of Matplotlib - this can be sometimes useful
* AutoViz automatically draws common graphs based on dataset

* Altair and others are new entrants that use proper JS based charting - truly interactive plots in the browser
* JS based charts do have a limitation on how much data they can handle (how much data your browser page can hold)
* There's a happy compromise somewhere in between

#### 'Backends'
Matplotlib can render graphs and charts using a variety of 'backends' - lowlevel code that actually do the drawing and painting. The backends can be accessed by the user facing plotting code. So we just setup a figure and use plt.show() and the backend does all the work of rendering the figure. This allows Matplotlib to address various needs - for e.g. rendering images, creating interactive widgets etc.

```python
# Create 50 evenly spaced numbers from 0 to 10
# Showing the last 20 values
x = np.linspace(0, 10, 50)
x[:]
```

## Plot Styling

```python
# Get the available styles
print(plt.style.available)
```

```python
# Plot the default matlplotlib style
# lots of curves - to see how colors work
plt.title('Default MatPlotLib')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(x, x**7.5)
plt.plot(x, x**7.75)
plt.plot(x, x**7.9)
plt.plot(x, x**8)
plt.plot(x, x**8.1)
plt.plot(x, x**8.25)
plt.show()
```

```python
# Plot the Bayesian Method for Hackers like style
plt.style.use('bmh')
plt.title('Bayesian Methods for Hackers')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(x, x**7.5)
plt.plot(x, x**7.75)
plt.plot(x, x**7.9)
plt.plot(x, x**8)
plt.plot(x, x**8.1)
plt.plot(x, x**8.25)
plt.show()
```

```python
# Plot the GGPlot like style
plt.style.use('ggplot')
plt.title('GGPlot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(x, x**7.5)
plt.plot(x, x**7.75)
plt.plot(x, x**7.9)
plt.plot(x, x**8)
plt.plot(x, x**8.1)
plt.plot(x, x**8.25)
plt.show()
```

```python
plt.style.use('grayscale')
plt.title('Grayscale')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(x, x**7.5)
plt.plot(x, x**7.75)
plt.plot(x, x**7.9)
plt.plot(x, x**8)
plt.plot(x, x**8.1)
plt.plot(x, x**8.25)
plt.show()
```

```python
plt.style.use('fivethirtyeight')
plt.title('Nate Silver fivethirtyeight.com')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(x, x**7.5)
plt.plot(x, x**7.75)
plt.plot(x, x**7.9)
plt.plot(x, x**8)
plt.plot(x, x**8.1)
plt.plot(x, x**8.25)
plt.show()
```

```python
plt.style.use('seaborn')
plt.title('seaborn')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(x, x**7.5)
plt.plot(x, x**7.75)
plt.plot(x, x**7.9)
plt.plot(x, x**8)
plt.plot(x, x**8.1)
plt.plot(x, x**8.25)
plt.show()
```

#### Bonus! XKCD Graphs!

```python
plt.style.use('classic')
plt.xkcd()
plt.title('XKCD Style Plots FTW!')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
# define a corkscrew
plt.plot(x, (np.sin(x*np.pi)*(x**1.5)))
plt.plot(x, (np.cos(x*np.pi+np.pi/2)*(x**1.5)))
plt.show()
```

#### Final notes on Matplotlib
There's a sense in which these style choices are for the final presentation/output.
For more exploratory use cases - interactive widgets may be a better idea:
* ipympl - ``%matplotlib ipympl`` or ``%matplotlib widgets``
* nbAgg - ``%matplotlib notebook`` - typically limited to one interactive figure

```python
# reset matplotlib styles
plt.style.use('default')
sns.lmplot(x='Attack', y='Defense', data = pokemon_df, hue='Stage')
# use fit_reg=false to remove the regression line
sns.lmplot(x='Attack', y='Defense', data = pokemon_df, hue='Stage', fit_reg = False)
```

```python
pokemon_stats_df = pokemon_df.drop(['Total', 'Stage', 'Legendary'], axis=1)
sns.boxplot(data=pokemon_stats_df)
```

```python
sns.violinplot(x='Type 1', y='Attack', data = pokemon_stats_df)
```

```python
sns.violinplot(x='Type 2', y='Attack', data = pokemon_stats_df)
```

```python
sns.swarmplot(x='Type 1', y='Attack', data=pokemon_stats_df)
```

```python
# overlay 2 plots together
plt.figure()
sns.violinplot(x='Type 1', y='Attack', data=pokemon_stats_df, inner=None) #remove the inner bar
sns.swarmplot(x='Type 1', y='Attack', data=pokemon_stats_df,
             color='k', alpha=0.5) #make points black and transparent
```
