#### 'Backends'
Matplotlib can render graphs and charts using a variety of 'backends' - lowlevel code that actually do the drawing and painting. The backends can be accessed by the user facing plotting code. So we just setup a figure and use plt.show() and the backend does all the work of rendering the figure. This allows Matplotlib to address various needs - for e.g. rendering images, creating interactive widgets etc.

#### Magics
The [Matplotlib magics](https://ipython.readthedocs.io/en/stable/interactive/magics.html#magic-matplotlib) are quick ways to configuring Matplotlib in a jupyter notebook.

## Data

```python
# Create 50 evenly spaced numbers from 0 to 10
# Showing the last 20 values
x = np.linspace(0, 10, 50)
x[:]
```

## Plot Styling

```python
# Get the available styles
print(plt.style.available)
```

```python
# Plot the default matlplotlib style
# lots of curves - to see how colors work
plt.title('Default MatPlotLib')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(x, x**7.5)
plt.plot(x, x**7.75)
plt.plot(x, x**7.9)
plt.plot(x, x**8)
plt.plot(x, x**8.1)
plt.plot(x, x**8.25)
plt.show()
```

```python
# Plot the default Baysian Methods for Hackers style
# plt.style.use('bmh')
plt.title('Baysian Methods for Hackers')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
# alternate way of specifying styles
with plt.style.context('bmh'):
    plt.plot(x, x**7.5)
    plt.plot(x, x**7.75)
    plt.plot(x, x**7.9)
    plt.plot(x, x**8)
    plt.plot(x, x**8.1)
    plt.plot(x, x**8.25)
plt.show()
```

```python
# Plot the GGPlot like style
plt.style.use('ggplot')
plt.title('GGPlot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(x, x**7.5)
plt.plot(x, x**7.75)
plt.plot(x, x**7.9)
plt.plot(x, x**8)
plt.plot(x, x**8.1)
plt.plot(x, x**8.25)
plt.show()
```

```python
plt.style.use('grayscale')
plt.title('Grayscale')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(x, x**7.5)
plt.plot(x, x**7.75)
plt.plot(x, x**7.9)
plt.plot(x, x**8)
plt.plot(x, x**8.1)
plt.plot(x, x**8.25)
plt.show()
```

```python
plt.style.use('fivethirtyeight')
plt.title('Nate Silver fivethirtyeight.com')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(x, x**7.5)
plt.plot(x, x**7.75)
plt.plot(x, x**7.9)
plt.plot(x, x**8)
plt.plot(x, x**8.1)
plt.plot(x, x**8.25)
plt.show()
```

```python
plt.style.use('seaborn')
plt.title('seaborn')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(x, x**7.5)
plt.plot(x, x**7.75)
plt.plot(x, x**7.9)
plt.plot(x, x**8)
plt.plot(x, x**8.1)
plt.plot(x, x**8.25)
plt.show()
```

```python
plt.style.use('seaborn-bright')
plt.title('seaborn-bright')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(x, x**7.5)
plt.plot(x, x**7.75)
plt.plot(x, x**7.9)
plt.plot(x, x**8)
plt.plot(x, x**8.1)
plt.plot(x, x**8.25)
plt.show()
```

```python
plt.style.use('seaborn-pastel')
plt.title('seaborn-pastel')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(x, x**7.5)
plt.plot(x, x**7.75)
plt.plot(x, x**7.9)
plt.plot(x, x**8)
plt.plot(x, x**8.1)
plt.plot(x, x**8.25)
plt.show()
```

```python
plt.style.use('dark_background')
plt.title('Dark Background')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(x, x**7.5)
plt.plot(x, x**7.75)
plt.plot(x, x**7.9)
plt.plot(x, x**8)
plt.plot(x, x**8.1)
plt.plot(x, x**8.25)
plt.show()
```

#### Final notes
There's a sense in which these style choices are for the final presentation/output.
For more exploratory use cases - interactive widgets may be a better idea:
* ipympl - ``%matplotlib ipympl`` or ``%matplotlib widgets``
* nbAgg - ``%matplotlib notebook`` - typically limited to one interactive figure
