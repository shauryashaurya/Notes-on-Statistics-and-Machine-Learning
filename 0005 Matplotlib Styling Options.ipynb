## Setup

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# turn on data table rendering
pd.set_option('display.notebook_repr_html', True)
```

## Data

```python
# Create 500 evenly spaced numbers from 0 to 10
# Showing the last 20 values
x = np.linspace(0, 10, 500)
x[-20:]
```

## Plot Styling

```python
# Get the available styles
print(plt.style.available)
```

```python
# Plot the default matlplotlib style
plt.title('Default MatPlotLib')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(x, x**8)
plt.show()
```

```python
# Plot the default Baysian Methods for Hackers style
plt.style.use('bmh')
plt.title('Baysian Methods for Hackers')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(x, x**8)
plt.show()
```

```python
# Plot the GGPlot like style
plt.style.use('ggplot')
plt.title('GGPlot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(x, x**8)
plt.show()
```

```python
plt.style.use('grayscale')
plt.title('Grayscale')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(x, x**8)
plt.show()
```

```python
plt.style.use('fivethirtyeight')
plt.title('Nate Silver fivethirtyeight.com')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(x, x**8)
plt.show()
```

```python
plt.style.use('dark_background')
plt.title('Dark Background')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(x, x**8)
plt.show()
```
